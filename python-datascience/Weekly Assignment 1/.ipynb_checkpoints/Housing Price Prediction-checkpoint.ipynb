```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

pd.set_option('display.max_columns', None)
```

```python
data_test = pd.read_csv('./test.csv')

data_train = pd.read_csv('./train.csv')
data_train.head()
```

```python
# assign data train to target
target = data_train['SalePrice']
target.head()
```

```python
data_train['FireplaceQu'].unique()
```

```python
# Check for missing data
missing_data = data_train.isnull().sum()
missing_data = missing_data[missing_data > 0].sort_values(ascending=False)
missing_data_percentage = (missing_data / len(data_train)) * 100

# Display missing data details
missing_data_df = pd.DataFrame({'Missing Values': missing_data, 'Percentage': missing_data_percentage})
missing_data_df
```

```python
# Define the threshold for dropping features with high percentage of missing data
missing_data_threshold = 80.0 # 80%

# Identify the features to drop (where the percentage of missing data is higher than the threshold)
features_to_drop = missing_data_percentage[missing_data_percentage > missing_data_threshold].index.tolist()

# Adding ID to drop
features_to_drop.append('Id')

# Drop the identified features from the dataset
data_train_reduced = data_train.drop(columns=features_to_drop)

# Show the features that were dropped
features_to_drop
```

```python
# Create a new category for missing values in 'MasVnrType' and 'FireplaceQu'
# Note: We checked 'MasVnrType' manually, and it had a high missing percentage as well
data_train_reduced['MasVnrType'].fillna('None', inplace=True)
data_train_reduced['FireplaceQu'].fillna('NoFireplace', inplace=True)
```

```python
numerical_features = data_train_reduced.select_dtypes(include=['int64', 'float64']).columns
numerical_missing = data_train_reduced[numerical_features].isnull().sum()
numerical_missing = numerical_missing[numerical_missing > 0].sort_values(ascending=False)

[numerical_missing, numerical_features]
```

```python
# https://vitalflux.com/pandas-impute-missing-values-mean-median-mode/#:~:text=Median%20imputation%20is%20preferred%20when,to%20outliers%20than%20the%20mean.
# Median imputation is preferred when the distribution is skewed, as the median is less sensitive to outliers than the mean.

for column in numerical_missing.index:
    median_value = data_train_reduced[column].median()
    data_train_reduced[column].fillna(median_value, inplace=True)

# Checking if there are any missing values left in numerical features
data_train_reduced[numerical_features].isnull().sum()
```

```python
# Identify categorical columns with missing values
categorical_features = data_train_reduced.select_dtypes(include=['object']).columns
categorical_missing = data_train_reduced[categorical_features].isnull().sum()
categorical_missing = categorical_missing[categorical_missing > 0].sort_values(ascending=False)

[categorical_features, categorical_missing]
```

```python
# https://medium.com/geekculture/frequent-category-imputation-missing-data-imputation-technique-4d7e2b33daf7

# Impute missing values in categorical features with the most frequent value (mode)
for column in categorical_missing.index:
    mode_value = data_train_reduced[column].mode()[0]
    data_train_reduced[column].fillna(mode_value, inplace=True)

# After imputation, check for any remaining missing values in categorical features
data_train_reduced[categorical_features].isnull().sum()
```
