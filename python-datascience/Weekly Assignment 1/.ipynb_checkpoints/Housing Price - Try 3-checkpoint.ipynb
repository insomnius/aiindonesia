```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split

pd.set_option('display.max_columns', None)
```

```python
housing_data = pd.read_csv('./train.csv')
housing_data.head()
```

```python
# Visualize the amount of missing data in each column as a bar chart
plt.figure(figsize=(12, 6))
missing = housing_data.isnull().sum()
missing = missing[missing > 0]
missing.sort_values(inplace=True)
missing.plot.bar()
plt.title('Number of Missing Values Per Column')
plt.xlabel('Columns')
plt.ylabel('Missing Values Count')
plt.show()
```

```python
# Create new categories for missing values that likely indicate the absence of a feature
housing_data['PoolQC'].fillna('No Pool', inplace=True)
housing_data['MiscFeature'].fillna('None', inplace=True)
housing_data['Alley'].fillna('No Alley', inplace=True)
housing_data['Fence'].fillna('No Fence', inplace=True)
housing_data['FireplaceQu'].fillna('No Fireplace', inplace=True)

# For 'LotFrontage', we'll impute with the median of the neighborhood's 'LotFrontage'
housing_data['LotFrontage'] = housing_data.groupby('Neighborhood')['LotFrontage'].transform(lambda x: x.fillna(x.median()))

# For Garage-related features, fill categorical ones with 'No Garage' and continuous with 0 or appropriate value
garage_features = ['GarageType', 'GarageFinish', 'GarageQual', 'GarageCond']
for g_feat in garage_features:
    housing_data[g_feat].fillna('No Garage', inplace=True)
housing_data['GarageYrBlt'].fillna(0, inplace=True)  # Assuming 0 represents no garage

# For Basement-related features, fill categorical ones with 'No Basement' and continuous with 0 or appropriate value
basement_features = ['BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2']
for b_feat in basement_features:
    housing_data[b_feat].fillna('No Basement', inplace=True)

# MasVnrType and MasVnrArea
housing_data['MasVnrType'].fillna('None', inplace=True)
housing_data['MasVnrArea'].fillna(0, inplace=True)

# Verify how many total missing values are left
total_missing_values = housing_data.isnull().sum().sum()
total_missing_values
```

```python
# Check which column still has missing values
missing_values = housing_data.isnull().sum()
remaining_missing = missing_values[missing_values > 0]
remaining_missing_column = remaining_missing.index[0]

# Since it's a single missing value, let's take a closer look at that column to determine the imputation strategy
column_data = housing_data[remaining_missing_column]
missing_value_info = {
    "Missing Values": remaining_missing.values[0],
    "Dtype": column_data.dtype,
    "Unique Values": column_data.unique()
}
missing_value_info
```

```python
# Impute the missing value with the mode of the column
mode_value = housing_data[remaining_missing_column].mode()[0]
housing_data[remaining_missing_column].fillna(mode_value, inplace=True)

# Re-check if there are any missing values left
total_missing_values_after_imputation = housing_data.isnull().sum().sum()
total_missing_values_after_imputation
```

```python
from sklearn.preprocessing import StandardScaler

# Identify categorical and numerical columns
categorical_cols = housing_data.select_dtypes(include=['object']).columns
numerical_cols = housing_data.select_dtypes(exclude=['object']).columns.drop(['Id', 'SalePrice'])

# Encode categorical features using one-hot encoding
encoded_data = pd.get_dummies(housing_data, columns=categorical_cols)

# Initialize the StandardScaler
scaler = StandardScaler()

# Scale the numerical features
encoded_data[numerical_cols] = scaler.fit_transform(encoded_data[numerical_cols])

# Now that the data is fully preprocessed, visualize the scaled numerical data
plt.figure(figsize=(12, 8))
sns.boxplot(data=encoded_data[numerical_cols])
plt.xticks(rotation=90)
plt.title('Boxplot of Scaled Numerical Features')
plt.show()
```
